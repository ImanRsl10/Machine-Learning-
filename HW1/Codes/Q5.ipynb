```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-4, 4, 100)
x2 = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1, x2)

w1_x = [-1.5,-0.5,0.5,1.5,1.5,0.5,1,-1,-1.5,-2]
w1_y = [1,0.5,0.5,0.5,-0.5,-0.5,-1,-1,0,-1]

w2_x = [0,1,1.5,2,2.5,1.5,0.5,1,2]
w2_y = [0.5,1,0,1,2,2,2,3,3]

Z = -0.638 * X1 ** 2 + 0.454 * X2 ** 2 + 0.3568 * X1 * X2 + 1.854 * X1 + 1.62 * X2 - 2.11

plt.scatter(w1_x, w1_y, label= 'class1', color= "black", marker= "o", s=30)
plt.scatter(w2_x, w2_y, label= 'class2', color= "red", marker= "*", s=40)

plt.contour(X1, X2, Z, levels=1, cmap='viridis')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Desicion Boundry')
plt.show()

Z_min_risk = -0.638 * X1 ** 2 + 0.454 * X2 ** 2 + 0.3568 * X1 * X2 + 1.854 * X1 + 1.62 * X2 - 1.81

plt.scatter(w1_x, w1_y, label= 'class1', color= "black", marker= "o", s=30)
plt.scatter(w2_x, w2_y, label= 'class2', color= "red", marker= "*", s=40)

plt.contour(X1, X2, Z_min_risk, levels=1, cmap='viridis')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Desicion Boundry')
plt.show()

Z_min_risk2 = -0.638 * X1 ** 2 + 0.454 * X2 ** 2 + 0.3568 * X1 * X2 + 1.854 * X1 + 1.62 * X2 - 1.11

plt.scatter(w1_x, w1_y, label= 'class1', color= "black", marker= "o", s=30)
plt.scatter(w2_x, w2_y, label= 'class2', color= "red", marker= "*", s=40)

plt.contour(X1, X2, Z_min_risk2, levels=1, cmap='viridis')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Desicion Boundry')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, a, b):
   return 1/np.pi * 1/(1 + pow((x - a)/b, 2))

x = np.arange(-10, 10, 0.01)

P_w1 = P_w2 = 0.5

P_w1_x = f(x, 3, 1) * P_w1 # P(w1|x) = P(x|w1) * P(w1)
P_w2_x = f(x, 5, 1) * P_w2

plt.plot(x, P_w1_x, color='r', label='P(w1|x)')
plt.plot(x, P_w2_x, color='b', label='P(w2|x)')
plt.xlabel("x")
plt.ylabel("Posterior")
plt.legend()
plt.show()

lambda12 = 1
lambda21 = 2

g_w1_x = P_w1_x * lambda21
g_w2_x = P_w2_x * lambda12

plt.plot(x, g_w1_x, color='r', label='g(w1|x)')
plt.plot(x, g_w2_x, color='b', label='g(w2|x)')
plt.xlabel("x")
plt.ylabel("Posterior")
plt.legend()
plt.show()
```

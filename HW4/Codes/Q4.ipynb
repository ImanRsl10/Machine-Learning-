```python
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt

# Sample data
X = np.array([[1, 4], [2, 3], [4, 5], [5, 6]])  # Replace with your data points
y = np.array([1, 1, -1, -1])  # Replace with your class labels

# Create an SVM classifier
clf = svm.SVC(kernel='linear')  # You can also use other kernels like 'poly', 'rbf', etc.

# Fit the model
clf.fit(X, y)

# Get the slope (w) and intercept (b)
w = clf.coef_[0]
b = clf.intercept_[0]

print(f"Slope (w): {w}")
print(f"Intercept (b): {b}")

# Plotting the data and decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')

# Plot the decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.xlabel("x1")
plt.ylabel("y1")
plt.title("SVM")

plt.show()
```
